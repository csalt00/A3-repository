```python
# defining the contact classes


class Contact:
  def __init__(self, name, age, degree, subject, fact):
    self.name = name
    self.age = age
    self.degree = degree
    self.subject = subject
    self.fact = fact

def __str__(self): #structure/pattern of return/display
    return f"Name: {self.name}, Age: {self.Age}, Degree: {self.degree}, Subject: {self.subject}, Fact: {self.fact}"

# creating contact contact book
class ContactBook:
  def __init__(self):
    self.contacts = {"name": "Alex",
      "age": "22", "degree":
      "Bachelor of Arts",
      "subject": "Media Influence",
      "fun_fact": "Hates Pizza"},{"name": "Morgan",
      "age": "20", "degree":
      "Bachelor of Sound Design", "subject": "WWDC",
      "fun_fact": "Allergic to bananas"}

    self.All_contacts = {}  #initialising the dictionary

  def add_contact(self, contact): #making the all contacts dynamic so inputs are added to the dictionary
    self.All_contacts[contact.name] = {"name": contact.name,
      "age": contact.age,
      "degree": contact.degree,
      "subject": contact.subject,
      "fact": contact.fact
      }


  def delete_contact(self, name): # function for deleting contacts
    for contact in self.contacts:
      if contact['name'] == name:
        self.contacts.remove(contact)
        if name in self.All_contacts:
          del self.All_contacts[name]
        return
    print(f"Contact with name '{name}' not found.")

  def search_contact(self, name): #function for searching contacts by name (can it be done by subject? stretch goal?)
    for contact in self.contacts:
      if contact['name'] == name:
         return contact
    return None

  def display_contacts(self): # function for displaying all contacts
    for contact in self.contacts:
      print(contact)

# Welcome screen (WIP stretch goal = make more User friendly/Visually pleasing)

def main():
  contact_book = ContactBook()

  while True: # contact home page
    print("\nContact Book Menu")
    print("1. Add contact")
    print("2. Delete contact")
    print("3. Search contact")
    print("4. Show all contacts")
    print("5. Quit")

# present menu to user

    choice = input("Enter your choice: ")

# creating instances of the above class - choices corresponding to user action e.g if user chooses 1 in welcome screen etc.

    if choice == "1":
      name = input("Enter contact's name: ")
      age = input("Enter contact's age: ")
      degree = input("Enter contact's degree: ")
      subject = input("Enter contact's subject: ")
      fact = input("Enter contact fact: ")
      contact = Contact(name, age, degree, subject, fact)
      contact_book.add_contact(contact)
      print("contact contact added :)")

      if contact not in contact_book.All_contacts:
        contact_book.All_contacts.update({contact.name})

      elif choice == "2":
        name = input("Enter the name of contact you want to delete: ")
        contact_book.delete_contact(name)
        print("Contact deleted. ")

#else if choice of 3 (searching contacts) search by class
      elif choice == "3":
          name = input("Enter the name to search:")
          contact = contact_book.search_contact(name)
      if contact:
        print("contact found:")
        print(contact)

      else:
          print("contact not found.")
# contact dictionary to display

      All_contacts = {}

    elif choice == "4":
                print("\nAll contacts: ")
                contact_book.display_contacts()

    elif choice == "5": # user journey if wanting to exit
                print("Goodbye!")

    else:
        print("Invalid choice. Please try again. ")

if __name__ == "__main__":
  main()

```
